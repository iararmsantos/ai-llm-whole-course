### Building AI Chatbots
- AI Assistant is very common
- Friendly, knowledgeable persona
- Ability to maintain context between messages
- Subject matter expertise to answer questions

The correct use of prompts
- The system prompt: set tone, establish ground-rules, provide critical background
- Context: insert context to give more relevant background information pertaining to the topic
- Multi-shot prompting: provide example conversations to prime for specific scenarios, train on conversational style and demonstrate complex iterations
