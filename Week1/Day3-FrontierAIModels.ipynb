### Base, Chat, and Reasoning Models: Understanding LLM Types
#### LLM comes in 3 flavors
  - Base Model: get input and predicts what comes next
  - Chat/Instruct Model: have a specific structure where it get a question and the model answer it
    - Better for interactive use cases and creative content(?) generation
  - Reasoning/Thinking: reason about the answer, decide how much reasoning, GPT has this flavor as well, so it is called hybrid.     
    - Budget forcing: You can add the word wait in the middle of the sequence then it wil make it to reasoning more
    - Better for problem solving

#### Frontier Models: GPT, Claude, Gemini & Their Strenghs and Pitfalls
  - Mind-blowing performance from Frontier LLMs:
    - Synthesizing Information
    - Fleshing out a skeleton: from simple notes build a well crafted email
    - Coding

  - Limitations of Frontier Models:
    - Specialized domains: not PhD level
    - Recent Events: limited knowledge
    - Can confidently make mistakes

  - Trying and comparing llms:
    - All are good synthesizing information and generating nuanced answers
    - Claude is more humorous, more attention to safety, more concise
    - innovations focused on lower cost variants
